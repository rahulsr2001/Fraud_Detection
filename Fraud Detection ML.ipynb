
# Insurance Fraud Detection Process

## 1.Data Collection
The insurance fraud detection dataset comprises various features related to insurance policies, claims, and policyholder information. Each record represents an insurance claim, and each attribute provides relevant details to help detect potential fraud.

## 2.Data Preprocessing
Before training models, data preprocessing is performed, including handling missing values, encoding categorical variables, scaling numerical features, and dividing the data into training and testing sets for evaluation.

## 3.Feature Selection
The following features are key to assessing the likelihood of fraudulent claims:

- **Month**: Month when the claim was made.
- **WeekOfMonth**: Week of the month for the claim event.
- **DayOfWeek**: Day of the week the claim was filed.
- **Make**: Vehicle make (brand).
- **AccidentArea**: Location type of the accident (Urban or Rural).
- **DayOfWeekClaimed**: Day of the week the claim was reported.
- **MonthClaimed**: Month when the claim was reported.
- **WeekOfMonthClaimed**: Week of the month the claim was reported.
- **Sex**: Gender of the policyholder.
- **MaritalStatus**: Marital status of the policyholder.
- **Age**: Age of the policyholder.
- **Fault**: Indicates whether the accident was the policyholder's fault.
- **PolicyType**: Type of policy held.
- **VehicleCategory**: Category of the vehicle.
- **FraudFound_P**: Target variable indicating if fraud was detected.
- **PolicyNumber**: Unique identifier for each policy.
- **RepNumber**: Representative handling the case.
- **Deductible**: Deductible amount for the policy.
- **DriverRating**: Rating assigned to the driver.
- **Days_Policy_Accident**: Days since policy inception to the accident.
- **Days_Policy_Claim**: Days between the accident and claim filing.
- **PastNumberOfClaims**: Number of prior claims.
- **AgeOfVehicle**: Age of the vehicle involved.
- **AgeOfPolicyHolder**: Age group of the policyholder.
- **PoliceReportFiled**: Whether a police report was filed.
- **WitnessPresent**: Indicates if there were witnesses.
- **AgentType**: Type of agent handling the case (Internal or External).
- **NumberOfSuppliments**: Number of supplemental reports.
- **AddressChange_Claim**: Time since the last address change.
- **NumberOfCars**: Number of cars owned.
- **Year**: Year of the policy.
- **BasePolicy**: Type of base policy (e.g., Liability, Collision).

## 4.Model Training
Several machine learning models are applied to classify claims as potentially fraudulent or not. The models and their respective accuracies are as follows:

- **Logistic Regression**: 94%
- **Support Vector Machine** (SVM): 94%
- **Decision Tree** (DT): 90%
- **K-Nearest Neighbors** (KNN): 94%
- **Random Forest** (RF): 94%
- **Naive Bayes** (NB): 93%
- **XGBoost** (XGB): 95%

## 5.Model Evaluation
Each model’s performance is assessed using accuracy scores, which represent the proportion of correct predictions on the test dataset. XGBoost achieved the highest accuracy (95%), making it the most suitable model for detecting fraud in this dataset.

## 6.Best Model Selection
Based on accuracy and reliability, XGBoost with an accuracy of 95% is selected as the best-performing model for insurance fraud detection.

## Conclusion
Through data analysis of policyholder details, vehicle information, and claim history, machine learning models effectively identify fraudulent claims. XGBoost provides the most accurate classification, helping to improve fraud detection in insurance claims.
